### Who scored more goals in total: Home Team or Away Team?

```python
home_team_goals = wc_2014['Home Team Goals'].sum()
away_team_goals = wc_2014['Away Team Goals'].sum()
[home_team_goals, away_team_goals]
```

### find number of total goals in World Cup 2014

```python
total_goals = home_team_goals + away_team_goals
total_goals
```

### determine the winner in each match

```python
def determineWinner(wc_2014):
    arr = []
    if wc_2014['Home Team Goals'] > wc_2014['Away Team Goals']:
        arr.append(wc_2014['Home Team Name'])
    elif wc_2014['Home Team Goals'] < wc_2014['Away Team Goals']:
        arr.append(wc_2014['Away Team Name'])
    else:
        winner = wc_2014['Win conditions'].split()
        if len(winner) != 0:
            arr.append(winner[0])
        else: arr.append('Draw')
    return arr
```

```python
winners = wc_2014.apply(determineWinner, axis=1)
wc_2014.insert(loc=9,column='Winner',value=winners)
wc_2014['WinnerStr'] = wc_2014['Winner'].apply(lambda x: ''.join(map(str,x)))
```

```python
wc_2014 = wc_2014.drop(['Winner','Winning Team', 'WinnerStr', 'win team'],axis=1)
```

### See how attendance changed from stage to stage

### Group by Stage

```python
groupped = wc_2014.groupby(by='Stage')
groupped_attendance = groupped.Attendance.mean()
groupped_attendance = groupped_attendance.reset_index()
groupped_attendance=groupped_attendance.apply(np.roll,shift=-1)
groupped_attendance
```

### Plot the attendance vs stage

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
sns.set()
```

```python
fig = plt.figure()
fig.subplots_adjust(top=1)
figure(figsize=(8, 6), dpi=80).suptitle('World Cup 2014: Attendance over Stages', fontsize='x-large', fontweight='bold')
sns.barplot(x="Stage", y="Attendance",
             data=groupped_attendance, palette = 'Dark2')
plt.xticks(rotation=45)
plt.legend(title='Attendance', labels=['Thousands of Fans'])
plt.savefig("output.png", dpi=300)
plt.show()
```

```python
fig = plt.figure()
fig.subplots_adjust(top=1)
figure(figsize=(8, 6), dpi=80).suptitle('World Cup 2014: Attendance over Stages', fontsize='x-large', fontweight='bold')
sns.lineplot(x="Stage", y="Attendance",
             data=groupped_attendance, palette = 'Dark2')
plt.xticks(rotation=45)
plt.legend(title='Attendance', labels=['Thousands of Fans'])
plt.savefig("output.png", dpi=300)
plt.show()
```
