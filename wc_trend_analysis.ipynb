```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wc_data = pd.read_csv('WorldCupMatches.csv')
```

```python
wc_data.columns
```

```python
groupped_data = wc_data.groupby(['Year'])
groupped_attendance = groupped_data['Attendance'].sum()
# df = groupped_attendance.to_frame()
col_names = ['Year','Attendance']
df = pd.DataFrame(groupped_attendance, columns=col_names)

years = groupped_attendance.reset_index()['Year']
attendance = groupped_attendance.reset_index()['Attendance']
groupped_attendance = groupped_attendance.reset_index()
```

```python
groupped_attendance
```

```python
groupped_data.sum()
df.name = 'Attendance1'
df1 = df['Attendance'].to_frame()
```

```python
# find total attendance every year and visualise it with matplotlib
```

```python
plt.plot(years, attendance, '-b.', label='Attendance')
plt.rcParams["figure.autolayout"] = True
plt.title('World Cup: Attendance over Years', fontsize=15)
plt.xlabel('Years', fontsize=15)
plt.ylabel('Attendance', fontsize=15)
plt.legend()
plt.margins(x=0, y=0)
plt.grid(True)
plt.show()
```

```python
# visualization with Seaborn
```

```python
import seaborn as sns
sns.set()
from matplotlib.pyplot import figure
```

```python
fig = plt.figure()
figure(figsize=(8, 6), dpi=80).suptitle('World Cup: Attendance over Years', fontsize='x-large', fontweight='bold')
fig.subplots_adjust(top=0.87)
sns.barplot(x="Year", y="Attendance",
             data=groupped_attendance, palette = 'Dark2')
plt.xticks(rotation=45)
plt.legend(title='Attendance', labels=['Thousands of Fans'])
# plt.savefig("output.png", dpi=300)
plt.show()
```

```python
fig = plt.figure()
figure(figsize=(8, 6), dpi=80).suptitle('World Cup: Attendance over Years', fontsize='x-large', fontweight='bold')
fig.subplots_adjust(top=0.87)
sns.lineplot(x="Year", y="Attendance",
             data=groupped_attendance, palette = 'Dark2')
plt.xticks(rotation=45)
plt.legend(title='Attendance', labels=['Thousands of Fans'])
# plt.savefig("output.png", dpi=300)
plt.show()
```
